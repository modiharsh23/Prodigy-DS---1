**Dropping unnecessary columns**

```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

**Extarction of top 10 countries with respect to total population**

```python
# Filter the DataFrame to include only rows where the Indicator Code is 'SP.POP.TOTL'
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']


# Sort the filtered data by the population in 2022 in descending order
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

# Get the top ten countries with the highest total population
total_top_ten_countries = total_population_sorted.head(10)

# Print the top ten countries, ensuring the column names are correct
print("Top ten countries with the highest total population in 2022\n")
print(total_top_ten_countries[['Country Code']])
```

**Bar Plot**

**Top Ten countries of total population in year 2022 and 2001**

```python
plt.figure(figsize=(15 ,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top ten Countries of total population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15 ,6))
plt.subplot(2,2,2)
sns.barplot(x="2001", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top ten Countries of total population (2001)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

### **Stacked Bar Plot**

**Top Ten countries with male and female population**

```python
merge_data=pd.merge(male_population_data,female_population_data,on='Country Code',suffixes=('_male', '_female'))
merge_data["Total population"] = merge_data["2022_male"] + merge_data["2022_female"]
```

```python
merge_data.head()
```

```python
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

**Top 10 countries with male and female population(2022)**

```python
# Get the top ten countries with the highest total population
top_ten_countries = sorted_data.head(10)
```

```python
plt.figure(figsize=(15 ,6))
sns.barplot(x="Country Code", y="2022_female", data=top_ten_countries, color="purple",label="female population")
sns.barplot(x="Country Code", y="2022_male", data=top_ten_countries, bottom=top_ten_countries["2022_female"], color="blue",label="Male population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

**Bottom 10 countries with male and female population(2022)**

```python
bottom_ten_countries = sorted_data.tail(10)
```

```python

plt.figure(figsize=(15, 6))

# Plotting female population
sns.barplot(x="Country Code", y="2022_female", data=bottom_ten_countries, color="yellow", label="Female population")

# Adding male population on top of female population using cumulative sum
sns.barplot(x="Country Code", y="2022_male", data=bottom_ten_countries, color="green", label="Male population")

plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
